Calcule a frequência absoluta para os sobreviventes no _Titanic_.

__Dica.:__ Utilize a função `.value_counts()`

Carregando as principais bibliotecas que iremos utilizar:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=sns.load_dataset('titanic')
```

### Questão 3

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Utilizando a coluna `Age` do _dataset Titanic_, defina as principais métricas estatísticas para essa variável tais como:
- Média;
- Desvio Padrão;
- Mínimo;
- Primeiro Quartil;
- Segundo Quartil (Mediana);
- Terceiro Quartil;
- Distância Interquartil (IQR);
- Máximo;
- Skewness;
- Moda.

```python
#1 jeito.
describe_age_df = df['age'].describe()
intervalo_interquartil = describe_age_df['75%'] - describe_age_df['25%']
#skewness(assimetria)
assimetria = df['age'].skew()
```

```python
# Segundo jeito.
age_data = df['age']

media = age_data.mean()
desvio_padrao = age_data.std()
minimo = age_data.min()
primeiro_quartil = age_data.quantile(0.25)
segundo_quartil = age_data.median()
terceiro_quartil = age_data.quantile(0.75)
iqr = terceiro_quartil - primeiro_quartil
maximo = age_data.max()
skewness = age_data.skew()
moda = age_data.mode()[0]

print("Média:", media)
print("Desvio Padrão:", desvio_padrao)
print("Mínimo:", minimo)
print("Primeiro Quartil:", primeiro_quartil)
print("Segundo Quartil (Mediana):", segundo_quartil)
print("Terceiro Quartil:", terceiro_quartil)
print("Distância Interquartil (IQR):", iqr)
print("Máximo:", maximo)
print("Skewness:", skewness)
print("Moda:", moda)
```

### Questão 4

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Crie um gráfico da dispersão das idades dos passageiros do _Titanic_ e identifique os pontos onde se encontram a média, mediana e moda das idades.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o dataset Titanic


# Extrair a coluna 'Age' do DataFrame
age_data = df['age'].dropna()  # Remover valores nulos

# Calcular média, mediana e moda
media = age_data.mean()
mediana = age_data.median()
moda = age_data.mode()[0]

# Criar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(age_data.index, age_data.values, alpha=0.5)

# Adicionar a média, mediana e moda ao gráfico
plt.axhline(y=media, color='r', linestyle='-', label=f'Média: {media:.2f}')
plt.axhline(y=mediana, color='g', linestyle='-', label=f'Mediana: {mediana}')
plt.axhline(y=moda, color='b', linestyle='-', label=f'Moda: {moda}')

# Adicionar rótulos e legenda
plt.xlabel('Índice do Passageiro')
plt.ylabel('Idade')
plt.title('Dispersão das Idades dos Passageiros do Titanic')
plt.legend()

# Exibir o gráfico
plt.grid(True)
plt.show()
```

### Questão 5

Os exercícios de 1 a 5 serão desenvolvidos utilizando o _dataset_ `Titanic.csv`:
    
Ainda trabalhando com os valores das idades crie três _Boxplots_:
- _Boxplot_ das idades para todos os passageiros;
- _Boxplot_ das idades dos passageiros separados pelo sexo.
- _Boxplot_ das idades dos passageiros separados pelo sexo e por sobreviventes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Configurações estéticas para os gráficos
sns.set(style="whitegrid")

# Boxplot das idades para todos os passageiros
plt.figure(figsize=(10, 6))
sns.boxplot(x='age', data=df, color='skyblue')
plt.title("Boxplot das Idades para Todos os Passageiros")
plt.xlabel("Idade")
plt.show()

# Boxplot das idades dos passageiros separados pelo sexo
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', data=df, palette='Set2')
plt.title("Boxplot das Idades dos Passageiros Separados pelo Sexo")
plt.xlabel("Sexo")
plt.ylabel("Idade")
plt.show()

# Boxplot das idades dos passageiros separados pelo sexo e por sobreviventes
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=df, palette='Set1')
plt.title("Boxplot das Idades dos Passageiros Separados pelo Sexo e por Sobreviventes")
plt.xlabel("Sexo")
plt.ylabel("Idade")
plt.legend(title="Sobreviventes", loc='upper right')
plt.show()
```
